```python
# color palette different

# sn.set(style="white")

def get_color_cycle(n=10):
    color_options = ['Set2'] 
#     color_options = ['Set1', 'Set2', 'Set3', 'Paired', 'Accent', 'Dark2']
    flatui = ["#9b59b6", "#3498db", "#e74c3c", "#2ecc71"]
#     random_color = random.choice(color_options)
#     p = sn.color_palette(flatui)
    p = sn.color_palette('Dark2')
    sn.set_palette(p)
    
    color_cycle = list()
    rainbow = plt.gca()._get_lines.color_cycle

    i = 0
    while i < n:
        c = rainbow.next()
        color_cycle.append(c)
        i += 1
#     random.shuffle(color_cycle)
    return iter(color_cycle)
```

```python
# human readable axis labels

def human_format(num, pos):
    magnitude = 0
    while abs(num) >= 1000:
        magnitude += 1
        num /= 1000.0
    # add more suffixes if you need them
    if round(num) == num:
        return '%d%s' % (num, ['', 'K', 'M', 'G', 'T', 'P'][magnitude])
    else:
        return '%.1f%s' % (num, ['', 'K', 'M', 'G', 'T', 'P'][magnitude])

def per_formatter(num, pos):
    if float(num) * 100 == round(float(num) * 100):
        return '%d%%' % (float(num) * 100)
    else:
        return '%.1f%%' % (float(num) * 100)

formatter = FuncFormatter(human_format)
per_formatter = FuncFormatter(per_formatter)
```
